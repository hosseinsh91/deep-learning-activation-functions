```python
import matplotlib.pyplot as plt
import numpy as np
```

# ELU

```python
def elu_function(x , alpha):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
def elu_derivative(x , alpha):
    return np.where(x >= 0, 1, alpha * np.exp(x))

x = np.arange(-10, 10 ,0.01)
alpha = 1

y = elu_function(x,alpha)
dy = elu_derivative(x,alpha)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label='ELU Function')
ax.plot(x, dy, label='ELU Derivative')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('ELU Activation Function and its Derivative')
ax.legend()
ax.set_xlim(-10, 10)
ax.set_ylim(-1, 8)
ax.xaxis.set_ticks(np.arange(-10, 11, 1))
ax.yaxis.set_ticks(np.arange(-2, 11, 1)) 
ax.axhline(0, color='black', linewidth=1) 
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
plt.show()
```

# RELU

```python
def relu_function(x):
    return np.where(x <= 0, 0, x)
def relu_derivative(x):
    return np.where(x <= 0, 0, 1)

x = np.arange(-10, 10 ,0.01)

y = relu_function(x)
dy = relu_derivative(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label='RELU Function')
ax.plot(x, dy, label='RELU Derivative')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('RELU Activation Function and its Derivative')
ax.legend()
ax.set_xlim(-10, 10)
ax.set_ylim(-1, 8)
ax.xaxis.set_ticks(np.arange(-10, 11, 1))
ax.yaxis.set_ticks(np.arange(-1, 11, 1)) 
ax.axhline(0, color='black', linewidth=1) 
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
plt.show()
```

# tanh

```python
def tanh_function(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def tanh_derivative(x):
    return 1- ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))**2

x = np.arange(-10, 10 ,0.01)

y = tanh_function(x)
dy = tanh_derivative(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label='tanh Function')
ax.plot(x, dy, label='tanh Derivative')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('tanh Activation Function and its Derivative')
ax.legend()
ax.set_xlim(-10, 10)
ax.set_ylim(-2, 2)
ax.xaxis.set_ticks(np.arange(-10, 11, 1))
ax.yaxis.set_ticks(np.arange(-2, 2, 0.5)) 
ax.axhline(0, color='black', linewidth=1) 
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
plt.show()
```

# Sigmoid

```python
def sigmoid_function(x):
    return 1 / (1+np.exp(-x))
def sigmoid_derivative(x):
    return (1 / (1+np.exp(-x))) * (1 - (1 / (1+np.exp(-x))))

x = np.arange(-10, 10 ,0.01)

y = sigmoid_function(x)
dy = sigmoid_derivative(x)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, label='sigmoid Function')
ax.plot(x, dy, label='sigmoid Derivative')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('sigmoid Activation Function and its Derivative')
ax.legend()
ax.set_xlim(-10, 10)
ax.set_ylim(-0.2, 1.2)
ax.xaxis.set_ticks(np.arange(-10, 11, 1))
ax.yaxis.set_ticks(np.arange(-0.2, 1.2, 0.1)) 
ax.axhline(0, color='black', linewidth=1) 
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
plt.show()
```

# SELU

```python
def selu_function(x , alpha , lamb):
    return np.where(x >= 0, x, lamb*alpha*(np.exp(x)-1))
def selu_derivative(x , alpha , lamb):
    return np.where(x >= 0, 1, lamb*alpha*np.exp(x))

x = np.arange(-10, 10 ,0.01)
alpha = 1.67326
lamb = 1.0507

y = selu_function(x , alpha , lamb)
dy = selu_derivative(x , alpha , lamb)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, label='SELU Function')
ax.plot(x, dy, label='SELU Derivative')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('SELU Activation Function and its Derivative')
ax.legend()
ax.set_xlim(-10, 10)
ax.set_ylim(-2, 5)
ax.xaxis.set_ticks(np.arange(-10, 11, 1))
ax.yaxis.set_ticks(np.arange(-2, 5, 0.5)) 
ax.axhline(0, color='black', linewidth=1) 
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
plt.show()
```
